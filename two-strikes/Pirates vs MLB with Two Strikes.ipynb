# Pirates vs MLB with Two Strikes

## Importing Statcast data as CSV files, initializing data structures

```python
import csv
import numpy as np
import matplotlib.pyplot as plt

evTwoStrikeDict, wobaTwoStrikeDict, isoTwoStrikeDict, baTwoStrikeDict = {}, {}, {}, {}
evDict, wobaDict, isoDict, baDict = {}, {}, {}, {}

piratesEvTwoStrikesDict, piratesWobaTwoStrikesDict, piratesIsoTwoStrikesDict, piratesBaTwoStrikesDict = {}, {}, {}, {}
piratesEvDict, piratesWobaDict, piratesIsoDict, piratesBaDict = {}, {}, {}, {}

ev, wobaMLBList, isoMLBList, baMLBList = [], [], [], []
ev2x, wobaTwoStrikesMLBList, isoTwoStrikesMLBList, baTwoStrikesMLBList = [], [], [], []

evPiratesPlayers, wobaPiratesList, isoPiratesList, baPiratesList = [], [], [], []
ev2xPirates, wobaTwoStrikesPiratesList, isoTwoStrikesPiratesList, baTwoStrikesPiratesList = [], [], [], []

evPirates = []
diffPiratesDict = {}

with open('2017_mlb_two_strikes.csv', 'rb') as csv2:
    reader2 = csv.reader(csv2, delimiter=',')
    reader2.next()
    for row in reader2:
        if (float(row[0]) > 25):
            evTwoStrikeDict[row[2]] = float(row[14])
            wobaTwoStrikeDict[row[2]] = float(row[9])
            isoTwoStrikeDict[row[2]] = float(row[8]) - float(row[5])
            baTwoStrikeDict[row[2]] = float(row[5])
```

```python
with open('2017_mlb_ahead.csv', 'rb') as csv1:
    reader = csv.reader(csv1, delimiter=',')
    reader.next()
    for row in reader:
        evDict[row[2]] = float(row[14])
        wobaDict[row[2]] = float(row[9])
        isoDict[row[2]] = float(row[8]) - float(row[5])
        baDict[row[2]] = float(row[5])
```

```python
with open('2017_pirates_two_strikes.csv', 'rb') as csv5:
    woba2xreader = csv.reader(csv5, delimiter=',')
    woba2xreader.next()
    for row in woba2xreader:
        piratesEvTwoStrikesDict[row[2]] = float(row[14])
        piratesWobaTwoStrikesDict[row[2]] = float(row[9])
        piratesIsoTwoStrikesDict[row[2]] = float(row[8]) - float(row[5])
        piratesBaTwoStrikesDict[row[2]] = float(row[5])
```

```python
with open('2017_pirates_ahead.csv', 'rb') as csv6:
    wobareader = csv.reader(csv6, delimiter=',')
    wobareader.next()
    for row in wobareader:
        if (float(row[0]) > 25):
            piratesEvDict[row[2]] = float(row[14])
            piratesWobaDict[row[2]] = float(row[9])
            piratesIsoDict[row[2]] = float(row[8]) - float(row[5])
            piratesBaDict[row[2]] = float(row[5])
```

```python
for player in piratesEvTwoStrikesDict.keys():
    if piratesEvDict.has_key(player):
        diffPiratesDict[player] = piratesEvDict[player] - piratesEvTwoStrikesDict[player]
        evPiratesPlayers.append(player)
        ev2xPirates.append(piratesEvTwoStrikesDict[player])
        evPirates.append(piratesEvDict[player])
        wobaTwoStrikesPiratesList.append(piratesWobaTwoStrikesDict[player])
        wobaPiratesList.append(piratesWobaDict[player])
        baTwoStrikesPiratesList.append(piratesBaTwoStrikesDict[player])
        baPiratesList.append(piratesBaDict[player])
        isoTwoStrikesPiratesList.append(piratesIsoTwoStrikesDict[player])
        isoPiratesList.append(piratesIsoDict[player])
```

```python
diffDict = {}
for player in evTwoStrikeDict.keys():
    if evDict.has_key(player):
        diffDict[player] = evDict[player] - evTwoStrikeDict[player]
        ev2x.append(evTwoStrikeDict[player])
        ev.append(evDict[player])
        isoTwoStrikesMLBList.append(isoTwoStrikeDict[player])
        isoMLBList.append(isoDict[player])
        wobaTwoStrikesMLBList.append(wobaTwoStrikeDict[player])
        wobaMLBList.append(wobaDict[player])
        baTwoStrikesMLBList.append(baTwoStrikeDict[player])
        baMLBList.append(baDict[player])
```

```python
print "Launch Speed Mean Differential: ", np.mean(diffDict.values())
print "Launch Speed Differential Standard Deviation: ", np.std(diffDict.values())
```

## MLB Launch Speed Differentials

```python
print "Differential","\t","Player"
for diff in sorted(diffDict, key=diffDict.get, reverse=True):
  print diffDict[diff], "\t\t", diff
```

### Pirates stats with and without 2 strikes (2x denotes with two strikes)

```python
print "{: <20}".format("Player"), "\t", "Diff", "\t", "wOBA", "\t", "2x", "\t", "ISO", "\t", "2x", "\t", "BA", "\t", "2x"
for diff in sorted(diffPiratesDict, key=diffPiratesDict.get, reverse=True):
  print "{: <20}".format(diff), "\t", diffPiratesDict[diff], "\t", piratesWobaDict[diff], "\t", piratesWobaTwoStrikesDict[diff], "\t", piratesIsoDict[diff], "\t", piratesIsoTwoStrikesDict[diff], "\t", piratesBaDict[diff], "\t", piratesBaTwoStrikesDict[diff]
```

```python
fig = plt.figure()
fig.suptitle('Batting Averages', fontsize=14, fontweight='bold')

ax = fig.add_subplot(111)
fig.subplots_adjust(top=0.85)
ax.set_title('With And Without Two Strikes (Min. 30 ABs)')

ax.set_xlabel('Without Two Strikes')
ax.set_ylabel('With Two Strikes')

mlb = ax.scatter(baMLBList, baTwoStrikesMLBList, color='y', zorder=1)
mlb.set_label('MLB')
z = np.polyfit(baMLBList, baTwoStrikesMLBList, 1)
p = np.poly1d(z)
ax.plot(baMLBList,p(baMLBList),"y", zorder = 2)

bucs = ax.scatter(baPiratesList, baTwoStrikesPiratesList, color='black', zorder=3)
bucs.set_label('Pirates')
zPirates = np.polyfit(baPiratesList, baTwoStrikesPiratesList, 1)
pPirates = np.poly1d(zPirates)
#ax.plot(baPiratesList, pPirates(baPiratesList), "black", zorder=4)

ax.annotate('Jose Osuna', xy=(0.366, 0.120), xytext=(0.4, 0.06),
            arrowprops=dict(facecolor='black', shrink=0.0),
            )

ax.annotate('Adam Frazier', xy=(0.324, 0.238), xytext=(0.15, 0.3),
            arrowprops=dict(facecolor='black', shrink=0.0),
            )
ax.legend()
plt.show()
#plt.savefig('ba_chart.pdf')

for player in evPiratesPlayers:
    print player, piratesBaDict[player], piratesBaTwoStrikesDict[player], p(piratesBaDict[player]), piratesBaTwoStrikesDict[player] - p(piratesWobaDict[player])
```

## MLB 2 Strike ISO Differences

```python
isoDiff = {}
for player in isoTwoStrikeDict.keys():
    if (isoDict.has_key(player)):
        isoDiff[player] = isoTwoStrikeDict[player] - p(isoDict[player])

print "{: <20}".format("Player"), "ISO - Expected ISO"
for diff in sorted(isoDiff, key=isoDiff.get, reverse=True):
  print "{: <20}".format(diff), isoDiff[diff]
```
